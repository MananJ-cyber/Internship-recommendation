```python
import math
import numpy as np 

def cosine_similarity(vec_a, vec_b):
    # 1. Calculate Dot Product using numpy's built-in function
    dot_product = np.dot(vec_a, vec_b)
    # 2. Calculate Magnitude of A
    mag_a = np.linalg.norm(vec_a)
    # 3. Calculate Magnitude of B
    mag_b = np.linalg.norm(vec_b)
    # 4. Return the formula result
    if mag_a == 0 or mag_b == 0:
        return 0.0
    return dot_product / (mag_a * mag_b)

# Test Data
vector_1 = np.array([1, 2, 3]) # Represents "Cat"
vector_2 = np.array([1, 2, 4]) # Represents "Kitten" (Should be high similarity)
vector_3 = np.array([9, 0, 1]) # Represents "Tractor" (Should be low similarity)

print('sim(vector_1, vector_2)=', cosine_similarity(vector_1, vector_2))
print('sim(vector_1, vector_3)=', cosine_similarity(vector_1, vector_3))
```

```python
import time
from google import genai
from google.genai import types

# 1. Initialize Client
client = genai.Client(api_key="YOUR_NEW_API_KEY") # Do NOT use the old exposed key

# 2. Upload Your File (The "Retrieval" part of RAG)
# You only need to do this once. The file is stored in Google's managed vector database.
file_path = "my_manual.pdf" 
print(f"Uploading {file_path}...")
uploaded_file = client.files.upload(path=file_path)

# Wait for processing to complete
while uploaded_file.state == "PROCESSING":
    print("Processing file...")
    time.sleep(2)
    uploaded_file = client.files.get(name=uploaded_file.name)

if uploaded_file.state != "ACTIVE":
    raise Exception(f"File failed to process: {uploaded_file.state}")

print("File ready for RAG!")

# 3. Configure the File Search Tool
# This tells Gemini: "Look inside this specific file to find answers."
file_search_tool = types.Tool(
    file_search=types.FileSearch() 
)

# 4. Call the Model
response = client.models.generate_content(
    model="gemini-2.5-pro", 
    contents="What are the safety warnings in this document?",
    config=types.GenerateContentConfig(
        tools=[file_search_tool], 
        # We attach the file specifically to this request context
        # This acts as the "Augmented" part of RAG
        documents=[uploaded_file] 
    )
)

print(response.text)

# Optional: See the citations (Where in the file it found the answer)
for candidate in response.candidates:
    for part in candidate.content.parts:
        if part.text:
            print(part.text)
```
